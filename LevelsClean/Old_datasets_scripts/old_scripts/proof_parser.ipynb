```python
def parse_lean_file(file_content):
    lines = file_content.split('\n')
    theorems = []
    
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        
        # Check if line starts a theorem
        if line.startswith('theorem '):
            theorem_data = {
                'theorem_statement': line,
                'theorem_NLs': [],
                'theorem_FLs': []
            }
            
            i += 1
            # Process lines until we hit the next theorem or end of file
            while i < len(lines) and not lines[i].strip().startswith('theorem '):
                line = lines[i].strip()
                
                # Skip empty lines
                if not line:
                    i += 1
                    continue
                
                # Check if line is a comment (NL)
                if line.startswith('--'):
                    theorem_data['theorem_NLs'].append(line)
                # Otherwise it's a formal language statement (FL)
                elif line:
                    theorem_data['theorem_FLs'].append(line)
                
                i += 1
            
            theorems.append(theorem_data)
        else:
            i += 1
    
    # Change the name of all theorems to "random_theorem"
    for theorem in theorems:
        theorem['theorem_NLs'] = theorem['theorem_NLs'][:-1]
        parts = theorem['theorem_statement'].split(' ')

        if len(parts) > 1:
            # Replace the theorem name (the second part) with "random_theorem"
            parts[1] = "random_theorem"
            theorem['theorem_statement'] = ' '.join(parts)
    
    
    
    return theorems

# Example usage:
# with open('your_lean_file.lean', 'r') as f:
#     file_content = f.read()
#     theorems = parse_lean_file(file_content)
#     for theorem in theorems:
#         print(f"Theorem: {theorem['theorem_statement']}")
#         print(f"NLs: {theorem['theorem_NLs']}")
#         print(f"FLs: {theorem['theorem_FLs']}")
#         print()

# Example usage:
with open('../Datasets/o1_test/test_1.lean', 'r') as f:
    file_content = f.read()
    theorems = parse_lean_file(file_content)
    for theorem in theorems:
        print(f"Theorem: {theorem['theorem_statement']}")
        print(f"NLs: {theorem['theorem_NLs']}")
        print(f"FLs: {theorem['theorem_FLs']}")
        print()
    print(len(theorems))

# save as json

import json
with open('../Datasets/o1_test/test_1.json', 'w') as f:
    json.dump(theorems,
              f,
              indent
              =4)

# load json
with open('../Datasets/o1_test/test_1.json', 'r') as f:
    theorems = json.load(f)
    for theorem in theorems:
        print(f"Theorem: {theorem['theorem_statement']}")
        print(f"NLs: {theorem['theorem_NLs']}")
        print(f"FLs: {theorem['theorem_FLs']}")
        print()
    print(len(theorems))
```
